# Query data dengan logical expression

## mengambil baris dengan 1 kondisi tertentu

```python
df.loc[df["order_id"] == "493410"]
```

```python
df.loc[df["order_id"] == "493410", ["product_name", "price", "customer_id"]]
```

```python
df.loc[(df["product_name"].str.contains("TEA")) & (df["quantity"] == 1), ["order_id", "product_name", "quantity"]]
```

```python
logic = (df["product_name"].str.contains("TEA")) | (df["product_name"].str.contains("Coffee"))

df.loc[logic]
```

```python
logic = (df["product_name"].str.contains("TEA")) | (df["product_name"].str.contains("Coffee") & (df["quantity"] >10))
df.loc[logic]
```
